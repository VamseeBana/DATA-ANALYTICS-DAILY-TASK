```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    "products":["curd_packets","sprite","bindu","lassy","milk","biscuits","choacaltes","cooldrinks"],
    "sales":[90,60,35,30,75,12,13,80],
    "product_stock":[100,80,50,50,80,40,40,100],
    "exipry_days":[7,20,20,10,7,30,30,20],
    "seanson":["all","sum","sum","sum","all","mid","all","sum"]

})
print(df)
```

```python
df1=df['product_stock']-df["sales"]
print(df1)
```

```python
df['avg_sales']=(df['sales']/df['product_stock']*100)
print(df)
```

```python
df2=df['avg_sales'].idxmax()
print(df['products'][df2])
```

FAST-MOVING PRODUCT IS MILK

```python
df3=df['avg_sales'].idxmin()
print(df['products'][df3])
```

SLOW-MOVING PRODUCT IS BISCUITS
