PREDICTIVE ANALYTICS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({
    "day":["monday","tuesday","wednesday","thursday","friday","sat","sun"],
    "waterconsumption":[160,170,180,190,200,210,220],
})
tankcapacitiy=220
```

```python
df1=df["waterconsumption"].mean()
print(df1)
```

```python
total_difference=df['waterconsumption'].diff()
print(total_difference)
print(total_difference.sum())
```

```python
avg=total_difference.sum()/7
print(avg)
```

```python
pedicion=avg+220
if pedicion >tankcapacitiy:
  print("shortage")
else:
  print(" no shotge")
```

```python
plt.bar(df["day"],df["waterconsumption"])
plt.show()
```
