```python
import pandas as pd
import numpy as np
df=pd.read_excel("/content/ap_weather_dataset.xlsx")
df.head()
```

```python
df.info()
```

```python
df_avg=df['Temp (°C)'].mean()
print(df_avg)
```

```python
df_avg1=df['Humidity'].mean()
print(df_avg1)
```

```python
df_avg2=df['Rainfall'].mean()
print(df_avg2)
```

```python
df_grop_temp=df.groupby('Location')["Temp (°C)"].mean()
print(df_grop_temp)
```

```python
df_greater_avg=df[df['Temp (°C)']>df_avg]
print(df_greater_avg)
```

```python
print(df_greater_avg['Location'].unique())
```

```python
df_greater_avg1=df[df['Humidity']>df_avg1]
print(df_greater_avg1)
```

```python
print(df_greater_avg1['Location'].unique())
```

📌 Insights
Insight 1

Coastal cities such as Visakhapatnam and Nellore have lower average temperatures compared to inland cities like Kurnool and Anantapur.

Insight 2

Locations with higher humidity show more moderate temperature conditions.

Insight 3

Higher rainfall during monsoon season reduces overall temperature.

Insight 4

Sea breeze and higher wind speed in coastal regions help maintain cooler temperatures.

Insight 5

Inland areas experience higher temperatures due to lower humidity and less coastal influence.

✅ Final Insight

The analysis indicates that coastal influence, rainfall, humidity, and wind speed are the major factors responsible for the moderate climate in Andhra Pradesh.
