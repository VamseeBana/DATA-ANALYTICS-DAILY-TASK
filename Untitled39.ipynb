DIAGONSTIC ANAYTICS

JIO USERS DATASET

```python
import pandas as pd
import numpy as np
import math
import statistics
df=pd.DataFrame(
    {
      "name":["vick","vamc","ram","rao","krishna","nani","nandu"],
      "intnet_speed":[200,700,80,400,300,100,390],
      "paymnt_failure":[3,0,5,1,2,1,0],
      "exta_costs":["yes","no","yes","no","yes","yes","no"],
      "srvices":["notsatisifid","satisifid","notsatisifid","satisifid","notsatisifid","notsatisifid","satisifid"]
    }
)
print(df)
```

LESS THAN AVERAGE SPEED

```python

df_intertnet_avg=df['intnet_speed'].mean()
print(df_intertnet_avg)

df_internet_spee_greater_avg=df[df['intnet_speed']<df_intertnet_avg]
print(df_internet_spee_greater_avg)
```

PAYEMENT FAILURES MORE THAN AVERAGE PAYEMENT

```python
df_payement_avg=df['paymnt_failure'].mean()
print(df_payement_avg)
df_payement_failure_greater_avg=df[df['paymnt_failure']>df_payement_avg]
print(df_payement_failure_greater_avg)
```

WHO ARE PAYING EXTRA COSTS

```python
extra = df[df['exta_costs'] == "yes"]
print(extra)
```

WHO  ARE COMMON IN ABOVE DATA

```python
df_group_inner_join = pd.merge(
    df_internet_spee_greater_avg,
    df_payement_failure_greater_avg,
    how='inner',
    on='name'
)

df_group_inner_join = pd.merge(
    df_group_inner_join,
    extra,
    how='inner',
    on='name'
)

#print(df_group_inner_join)
print(df_group_inner_join['name'])
```
